```python
dff1=pd.read_csv('FMCG_2022_2024.csv')
df1=dff1
```

```python
df1.head()
```

```python
df1['date'] = pd.to_datetime(df1['date'])
df1=df1[df1['date']<pd.Timestamp('2024-12-01')]
```

```python
general=df1.groupby(['date']).agg({'ventas2':'sum'}).reset_index()
df_train=general
```

```python
df1['date'] = pd.to_datetime(df1['date'])
general=df1.groupby(['date']).agg({'ventas2':'sum'}).reset_index()
df_train=general
```

```python
df_train=df_train.rename(columns={'date':'ds','ventas2':'y'})
```

```python
df_train.tail(50)
```

```python
df_prophet=df_train.rename(columns={'FECHA':'ds','VENTA_NETA':'y'})
modelo = Prophet(interval_width=0.95)
modelo.fit(df_prophet)
futuro = modelo.make_future_dataframe(periods=35)
prediccion = modelo.predict(futuro)
modelo.plot(prediccion)

```

```python
prediccion=prediccion[['ds','yhat','yhat_lower','yhat_upper']]
```
